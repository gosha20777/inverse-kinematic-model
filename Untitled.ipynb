# Inverse Kinematic

### 0. Imports

```python
import os
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
import pandas as pd
from keras import backend as K
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from random import randint
import numpy as np
from requests import get
```

### 1. Download Datasets

```python
def download(url, file_name):
    with open(file_name, "wb") as file:
        response = get(url)
        file.write(response.content)
if not os.path.exists('data'):
        os.makedirs('data')
        
#download('url','data/up_body.csv')
#download('url','data/up_body_test.csv')
#download('url','data/down_body.csv')
#download('url','data/down_body_test.csv')
```

### 2. Up Body
Prepare

```python
# Load data
df = pd.read_csv("data\BodyUp.csv", sep=";")
dataset = df.iloc[1:,:]
print(dataset.shape)
## UNNORMALIZED
Y = df.iloc[:,[34, 35, 36]]
X = df.iloc[:,[19, 20, 21, 22, 23, 24]]
print(X)
print(Y)

df_test = pd.read_csv("data\BodyUpTest.csv", sep=";")
## UNNORMALIZED
Y_test = df_test.iloc[:,[34, 35, 36]]
X_test = df_test.iloc[:,[19, 20, 21, 22, 23, 24]]
print(X.shape,Y.shape)
print(X_test.shape,Y_test.shape)

def base_model():
     model = Sequential()
     model.add(Dense(32, input_dim=6, init='normal', activation='relu'))
     model.add(Dense(64, init='normal', activation='relu'))
     model.add(Dense(128, init='normal', activation='relu'))
     model.add(Dense(512, init='normal', activation='relu'))
     model.add(Dense(128, init='normal', activation='relu'))
     model.add(Dense(64, init='normal', activation='relu'))
     model.add(Dense(32, init='normal', activation='relu'))
     model.add(Dense(3, init='normal'))
     model.compile(loss='mean_absolute_error', optimizer = 'adam')
     return model
    
clf = KerasRegressor(build_fn=base_model, epochs=120, batch_size=32,verbose=2)
#KerasRegressor(build_fn=base_model, epochs=120000, batch_size=32,verbose=2)
```

Train

```python
clf.fit(X,Y)
```

Predict and make tests

```python
def query(count, x, y):
    samples = len(x)
    for i in range (0, count):
        print('---TEST #' + str(i+1) + '---')
        index = randint(0, samples-2)
        _x = np.array(x)
        _y = np.array(y)
        print('Sample: ' + str(index))
        print('Query: ' + str(_x[index]))
        print('True: ' + str(_y[index]))
        prediction = clf.predict(np.array([_x[index]]))
        print('Prediction: ' + str(prediction))
        
query(10, X_test, Y_test)

res = clf.predict(X_test)
score = mean_absolute_error(Y_test, res)
print("unnormalized_absolute_error: %.2f" % (score))
print("real_absolute_error: %.2f" % (score / 8.52)) # scale size to normalzed
score = mean_squared_error(Y_test, res)
print("unnormalized_squared_error: %.2f MSE" % (score))
print("real_squared_error: %.2f MSE" % (score / 8.52))
score = r2_score(Y_test, res)
print("r2_score: %.2f MSE" % (score))
```
